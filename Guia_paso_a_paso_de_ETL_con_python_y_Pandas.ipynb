## ¿Qué es ETL?
ETL es un proceso fundamental en la gestión de datos que implica tres fases principales:
*	**Extracción (Extract):** Consiste en obtener los datos de diversas fuentes, como bases de datos, archivos planos, APIs, etc.
*	**Transformación (Transform):** Los datos extraídos son limpiados, enriquecidos y estructurados para que sean adecuados para el análisis o el almacenamiento en el destino final. Esto puede incluir la eliminación de duplicados, la corrección de errores, la normalización de formatos y la agregación de información.
*	**Carga (Load):** Los datos transformados se cargan en el sistema de destino, que puede ser un data warehouse, una base de datos analítica o cualquier otro sistema de almacenamiento para su posterior uso.

## Paso 2: Crear Archivos .csv de Muestra

Para este ejemplo, crearemos dos archivos .csv de muestra que simularemos extraer. Los crearemos directamente en el entorno de Colab.

En una celda de tu notebook de Colab, ejecuta el siguiente código:

```python
import pandas as pd

# Datos para el primer CSV: 'ventas.csv'
data_ventas = {
    'id_venta': [1, 2, 3, 4, 5, 6],
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Laptop', 'Webcam'],
    'cantidad': [1, 2, 1, 1, 1, 3],
    'precio_unitario': [1200.00, 25.00, 75.00, 300.00, 1200.00, 40.00],
    'fecha_venta': ['2024-01-15', '2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', '2024-01-19']
}
df_ventas = pd.DataFrame(data_ventas)
df_ventas.to_csv('ventas.csv', index=False)

# Datos para el segundo CSV: 'clientes.csv'
data_clientes = {
    'id_cliente': [101, 102, 103, 104, 105, 106],
    'nombre_cliente': ['Ana García', 'Luis Pérez', 'María López', 'Juan Rodríguez', 'Ana García', 'Pedro Díaz'],
    'ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia', 'Madrid', 'Bilbao'],
    'email': ['ana.g@example.com', 'luis.p@example.com', 'maria.l@example.com', 'juan.r@example.com', 'ana.g@example.com', 'pedro.d@example.com']
}
df_clientes = pd.DataFrame(data_clientes)
df_clientes.to_csv('clientes.csv', index=False)

print("Archivos 'ventas.csv' y 'clientes.csv' creados exitosamente en Google Colab.")
```

# Proceso ETL con Python y Pandas

Ahora, vamos a implementar el proceso ETL paso a paso.



# Paso 1: Extracción (Extract)

En esta fase, leeremos los archivos .csv que acabamos de crear en DataFrames de Pandas.

En una nueva celda de Colab, ejecuta el siguiente código:

```python
import pandas as pd

# Extracción de datos de 'ventas.csv'
try:
    df_ventas = pd.read_csv('ventas.csv')
    print("DataFrame 'df_ventas' (Extracción de ventas):")
    print(df_ventas.head())
    print("\nColumnas de df_ventas:", df_ventas.columns.tolist())
    print("\nTipos de datos de df_ventas:\n", df_ventas.dtypes)
except FileNotFoundError:
    print("Error: 'ventas.csv' no encontrado. Asegúrate de haberlo creado.")

print("-" * 30)

# Extracción de datos de 'clientes.csv'
try:
    df_clientes = pd.read_csv('clientes.csv')
    print("\nDataFrame 'df_clientes' (Extracción de clientes):")
    print(df_clientes.head())
    print("\nColumnas de df_clientes:", df_clientes.columns.tolist())
    print("\nTipos de datos de df_clientes:\n", df_clientes.dtypes)
except FileNotFoundError:
    print("Error: 'clientes.csv' no encontrado. Asegúrate de haberlo creado.")
```

## Explicación:

*   Usamos `pd.read_csv()` para cargar los datos de los archivos CSV en DataFrames.
*  `df.head()` muestra las primeras filas del DataFrame.
*   `df.columns.tolist()` muestra los nombres de las columnas.
*  `df.dtypes` muestra los tipos de datos de cada columna.



# Paso 2: Transformación (Transform)

Aquí realizaremos varias operaciones de limpieza y enriquecimiento de datos.

En una nueva celda de Colab, ejecuta el siguiente código:

```python
# Asegurarse de que los DataFrames existen después de la extracción
if 'df_ventas' in locals() and 'df_clientes' in locals():
    print("Iniciando la fase de Transformación...")

    # Transformación 1: Calcular el total de venta en df_ventas
    df_ventas['total_venta'] = df_ventas['cantidad'] * df_ventas['precio_unitario']
    print("\nDataFrame 'df_ventas' después de calcular 'total_venta':")
    print(df_ventas.head())

    # Transformación 2: Convertir 'fecha_venta' a tipo datetime
    df_ventas['fecha_venta'] = pd.to_datetime(df_ventas['fecha_venta'])
    print("\nTipos de datos de df_ventas después de convertir 'fecha_venta':\n", df_ventas.dtypes)

    # Transformación 3: Eliminar duplicados en df_ventas (si los hubiera, basándonos en producto, cantidad y fecha)
    # Por ejemplo, si una venta idéntica se registró dos veces por error
    initial_rows_ventas = df_ventas.shape[0]
    df_ventas.drop_duplicates(subset=['producto', 'cantidad', 'precio_unitario', 'fecha_venta'], inplace=True)
    if df_ventas.shape[0] < initial_rows_ventas:
        print(f"\nSe eliminaron {initial_rows_ventas - df_ventas.shape[0]} duplicados en df_ventas.")
    else:
        print("\nNo se encontraron duplicados significativos en df_ventas para eliminar.")

    # Transformación 4: Unir (Merge) df_ventas con df_clientes
    # Asumimos que podemos unir por 'id_cliente' y que 'id_venta' en df_ventas puede estar relacionado con un cliente
    # Para este ejemplo, simularemos una relación, ya que los CSVs no la tienen explícitamente en el 'id_venta'.
    # Si 'id_venta' en df_ventas fuera 'id_cliente', la unión sería directa.
    # Para propósitos demostrativos, agregaremos una columna 'id_cliente' a df_ventas para la unión.
    # En un escenario real, 'id_cliente' sería parte de los datos de ventas.
    df_ventas['id_cliente'] = [101, 102, 103, 104, 105, 106] # Asignación ficticia para el merge

    df_merged = pd.merge(df_ventas, df_clientes, left_on='id_cliente', right_on='id_cliente', how='left')
    print("\nDataFrame 'df_merged' (ventas con información de clientes):")
    print(df_merged.head())
    print("\nColumnas de df_merged:", df_merged.columns.tolist())

    # Transformación 5: Agregación - Total de ventas por cliente y ciudad
    df_ventas_por_cliente_ciudad = df_merged.groupby(['nombre_cliente', 'ciudad'])['total_venta'].sum().reset_index()
    df_ventas_por_cliente_ciudad.rename(columns={'total_venta': 'total_gastado'}, inplace=True)
    print("\nDataFrame 'df_ventas_por_cliente_ciudad' (Total gastado por cliente y ciudad):")
    print(df_ventas_por_cliente_ciudad)

    # Transformación 6: Limpiar la columna 'email' (ej. convertir a minúsculas)
    df_merged['email'] = df_merged['email'].str.lower()
    print("\nColumna 'email' en df_merged después de convertir a minúsculas:")
    print(df_merged[['nombre_cliente', 'email']].head())

else:
    print("Los DataFrames 'df_ventas' o 'df_clientes' no existen. Asegúrate de ejecutar la fase de Extracción primero.")
```

## Explicación de las transformaciones:


* **Cálculo de** `total_venta`**:** Se crea una nueva columna multiplicando la cantidad por el precio unitario.

* **Conversión de tipo de dato:** La columna fecha_venta se convierte a un formato de fecha y hora (datetime) para facilitar operaciones basadas en tiempo.

* **Eliminación de duplicados:** Se buscan y eliminan filas duplicadas basándose en un subconjunto de columnas para asegurar la unicidad de los registros de ventas.

* **Unión (Merge) de DataFrames:** Se combinan los DataFrames de ventas y clientes utilizando la columna id_cliente (simulada para este ejemplo) para enriquecer los datos de ventas con información del cliente.

* **Agregación:** Se agrupan los datos por cliente y ciudad para calcular el total de ventas por cada combinación.

* **Limpieza de texto:** La columna email se convierte a minúsculas para estandarizar los datos.

# Paso 3: Carga (Load)

Finalmente, cargaremos los datos transformados en un nuevo archivo .csv.

En una nueva celda de Colab, ejecuta el siguiente código:



```python
# Asegurarse de que el DataFrame 'df_merged' exista después de la transformación
if 'df_merged' in locals():
    # Carga: Guardar el DataFrame transformado en un nuevo archivo CSV
    output_file_path = 'ventas_transformadas.csv'
    df_merged.to_csv(output_file_path, index=False)
    print(f"\nDataFrame transformado cargado exitosamente en '{output_file_path}'.")

    # Carga: Guardar el DataFrame agregado en otro archivo CSV
    output_agg_file_path = 'ventas_agregadas_por_cliente.csv'
    df_ventas_por_cliente_ciudad.to_csv(output_agg_file_path, index=False)
    print(f"DataFrame agregado cargado exitosamente en '{output_agg_file_path}'.")

    # Opcional: Mostrar las primeras filas del archivo cargado para verificar
    print("\nVerificando el archivo 'ventas_transformadas.csv':")
    df_loaded_transformed = pd.read_csv(output_file_path)
    print(df_loaded_transformed.head())

    print("\nVerificando el archivo 'ventas_agregadas_por_cliente.csv':")
    df_loaded_agg = pd.read_csv(output_agg_file_path)
    print(df_loaded_agg.head())

else:
    print("El DataFrame 'df_merged' o 'df_ventas_por_cliente_ciudad' no existen. Asegúrate de ejecutar la fase de Transformación primero.")
```

## Explicación:

* `df.to_csv()` se utiliza para guardar el DataFrame transformado en un nuevo archivo CSV.

* `index=False` evita que Pandas escriba el índice del DataFrame como una columna en el CSV.

* Se guardan dos archivos: uno con los datos de ventas transformados y otro con los datos agregados por cliente.
